```julia
function shiftmap(x)
    return 2*x % 1   
end
```

```julia
shiftmap(1/3)
```

```julia
function iteratemap(x0,n::Int64, mapname=shiftmap)
    x=Array{typeof(x0),1}(undef,n)
    for i in 1:n
        x0 = mapname(x0)
        x[i] = x0
    end
    return x
end
```

```julia
x=iteratemap(1/3,100,shiftmap)
```

```julia
using Plots
```

```julia
plot(x)
```

```julia
plot(iteratemap(1709/5077,100,shiftmap))
```
